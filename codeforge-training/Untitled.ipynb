```python
import json

training_data = []

# Example 1 — Porter's Five Forces on Assets gap
training_data.append({
    "instruction": "Analyse this manufacturing company's KPI data using Porter's Five Forces framework",
    "input": """Industry: Float Glass Manufacturing
Asset Score: 64% | Outcome Score: 54%
Top Asset Gaps:
- Employee Attrition Rate: 42% score (Actual: 12%, Target: 5%)
- PPE Compliance: 85% score
- Machine Downtime: 64% score

Top Outcome Gaps:
- Digital Orders per month: 16% score (Actual: 80, Target: 500)
- On-Time Delivery: 72% score""",

    "output": """Porter's Five Forces Analysis — Float Glass Manufacturing

FORCE 1: COMPETITIVE RIVALRY — HIGH RISK 🔴
Your overall score of 59% indicates you are underperforming vs likely competitors.
Machine downtime at 64% score means your production efficiency is below par.
Competitors operating at higher efficiency can undercut your pricing and delivery times.
Priority: Reduce machine downtime to below 10% to restore competitive parity.

FORCE 2: THREAT OF NEW ENTRANTS — MEDIUM ⚠️  
Your PPE compliance at 85% and attrition at 42% score suggest operational vulnerabilities.
New entrants with modern facilities and lower attrition can enter and quickly match output.
Your digital orders gap (16% score) means you have no strong digital moat.
Priority: Build digital channel strength before new entrants capture that market.

FORCE 3: BARGAINING POWER OF SUPPLIERS — HIGH RISK 🔴
Attrition rate at 42% score is your biggest supplier risk — skilled labour is your key input.
At 12% actual attrition vs 5% target, you are losing 7% more staff than sustainable.
This forces dependency on recruitment agencies and training pipelines, increasing costs.
Priority: Invest in retention programs immediately — this is your most critical gap.

FORCE 4: BARGAINING POWER OF BUYERS — MEDIUM ⚠️
On-time delivery at 72% gives buyers leverage to negotiate or switch suppliers.
Digital orders at 16% means buyers cannot self-serve — high friction reduces loyalty.
Priority: Improve delivery reliability to above 90% to reduce buyer switching intent.

FORCE 5: THREAT OF SUBSTITUTES — LOW-MEDIUM 🟡
Float glass has limited direct substitutes in construction and automotive sectors.
However your digital channel weakness (16%) exposes you to online-first competitors.
Priority: Establish digital ordering capability to defend against platform substitutes.

OVERALL STRATEGIC RECOMMENDATION:
Your two most urgent actions based on Porter's analysis:
1. Employee retention programme — addresses Forces 1, 2, and 3 simultaneously
2. Digital channel investment — addresses Forces 2, 4, and 5 simultaneously
These two actions alone would close approximately 40% of your total performance gap."""
})

print(f"Training examples so far: {len(training_data)}")
print("First example added ✅")
```

```python
# Example 2 — High performing company
training_data.append({
    "instruction": "Analyse this manufacturing company's KPI data using Porter's Five Forces framework",
    "input": """Industry: Float Glass Manufacturing
Asset Score: 82% | Outcome Score: 78%
Top Asset Gaps:
- Employee Attrition Rate: 88% score (Actual: 5.5%, Target: 5%)
- Machine Downtime: 76% score (Actual: 9.5%, Target: 7%)
- Energy Consumption: 71% score

Top Outcome Gaps:
- Digital Orders: 71% score (Actual: 355, Target: 500)
- Customer Satisfaction: 85% score
- On-Time Delivery: 91% score""",

    "output": """Porter's Five Forces Analysis — Float Glass Manufacturing

FORCE 1: COMPETITIVE RIVALRY — LOW-MEDIUM 🟢
Your overall score of 80% places you in a strong competitive position.
Machine downtime at 76% and delivery at 91% show solid operational performance.
You are likely outperforming most regional competitors on reliability.
Priority: Close the energy consumption gap to further reduce costs vs rivals.

FORCE 2: THREAT OF NEW ENTRANTS — LOW 🟢
Strong attrition score (88%) means you retain skilled workers competitors cannot easily poach.
High delivery reliability creates customer loyalty that new entrants struggle to break.
Your main vulnerability is digital orders at 71% — new digital-first entrants could threaten this.
Priority: Accelerate digital channel to build a moat before new entrants exploit the gap.

FORCE 3: BARGAINING POWER OF SUPPLIERS — LOW 🟢
Attrition at 88% score means low dependency on external recruitment — strong position.
You control your workforce pipeline well reducing supplier leverage.
Priority: Maintain current retention levels — this is a genuine competitive advantage.

FORCE 4: BARGAINING POWER OF BUYERS — LOW 🟢
On-time delivery at 91% gives buyers little reason to seek alternatives.
Customer satisfaction at 85% indicates strong loyalty and low switching intent.
Digital orders at 71% is the only gap buyers could use as leverage.
Priority: Bring digital ordering to 90%+ to eliminate last buyer leverage point.

FORCE 5: THREAT OF SUBSTITUTES — LOW 🟢
Your strong delivery and satisfaction scores make switching to substitutes unattractive.
Energy consumption gap (71%) could become a risk if sustainability regulations tighten.
Priority: Address energy efficiency now before it becomes a compliance-driven substitute threat.

OVERALL STRATEGIC RECOMMENDATION:
You are in a strong position across all five forces.
Your single biggest strategic opportunity:
1. Digital channel (71%) — the only significant gap remaining
2. Energy efficiency (71%) — future-proofs against regulatory substitutes
Focus here will move you from strong to market-leading position."""
})

# Example 3 — Critical situation
training_data.append({
    "instruction": "Analyse this manufacturing company's KPI data using Porter's Five Forces framework",
    "input": """Industry: Float Glass Manufacturing
Asset Score: 38% | Outcome Score: 31%
Top Asset Gaps:
- Employee Attrition Rate: 22% score (Actual: 23%, Target: 5%)
- Machine Downtime: 31% score (Actual: 32%, Target: 10%)
- Safety Incidents: 28% score

Top Outcome Gaps:
- Revenue Growth: 18% score (Actual: 2%, Target: 11%)
- Digital Orders: 12% score (Actual: 60, Target: 500)
- Customer Complaints: 35% score""",

    "output": """Porter's Five Forces Analysis — Float Glass Manufacturing

FORCE 1: COMPETITIVE RIVALRY — CRITICAL 🔴🔴
Overall score of 35% indicates severe competitive disadvantage.
Machine downtime at 31% means 32% of production time is lost — competitors producing 3x more.
Revenue growth at 18% score confirms you are losing market share actively right now.
URGENT: Machine reliability must be treated as a business survival issue, not a KPI target.

FORCE 2: THREAT OF NEW ENTRANTS — CRITICAL 🔴🔴
At 35% overall, new entrants need only basic competence to outperform you.
Your digital channel at 12% score means you have almost no online presence.
Safety incidents at 28% score create reputational risk that accelerates customer loss.
URGENT: Any new entrant in your market is a serious threat at your current performance level.

FORCE 3: BARGAINING POWER OF SUPPLIERS — CRITICAL 🔴🔴
Attrition at 22% score means 23% of your workforce leaves annually — nearly 1 in 4 people.
You are completely dependent on constant recruitment and training pipelines.
Recruitment agencies have maximum leverage over you — high cost, low quality control.
URGENT: Workforce instability is feeding every other gap — this must be fixed first.

FORCE 4: BARGAINING POWER OF BUYERS — CRITICAL 🔴🔴
Customer complaints at 35% score combined with delivery issues gives buyers full leverage.
Buyers can and likely are switching to competitors right now.
Revenue growth at 18% score confirms buyer attrition is already happening.
URGENT: Conduct emergency customer retention outreach before further revenue loss.

FORCE 5: THREAT OF SUBSTITUTES — HIGH 🔴
At this performance level any alternative supplier is effectively a substitute.
Digital gap at 12% means online platforms can easily position as a better alternative.
URGENT: Minimum viable digital presence needed immediately to stop losing online buyers.

OVERALL STRATEGIC RECOMMENDATION:
This business is in a critical situation requiring immediate intervention.
Prioritise in this exact order:
1. Machine downtime — losing 32% production is unsustainable financially
2. Employee attrition — 23% annual turnover is destroying every other metric
3. Customer retention — contact top 10 customers this week before they switch
4. Digital presence — minimum viable online ordering within 60 days
Without urgent action on points 1 and 2, the other gaps cannot be closed."""
})

print(f"Training examples so far: {len(training_data)}")
print("Examples 2 and 3 added ✅")
```
