```python
# Cell 2: Imports
import json
import random
import pandas as pd
from datasets import load_dataset
from tqdm import tqdm

random.seed(42)
print('Imports done')
```

```python
# Cell 3: Download public coding datasets (Option A — General Coding)
print('Downloading CodeAlpaca-20k...')
alpaca = load_dataset('sahil2801/CodeAlpaca-20k', split='train')
print(f'CodeAlpaca: {len(alpaca)} examples')

# Convert to list of dicts
public_data = []
for item in alpaca:
    if item.get('output') and len(item['output'].strip()) > 20:
        public_data.append({
            'instruction': item['instruction'],
            'input': item.get('input', ''),
            'output': item['output'],
            'source': 'codealpaca'
        })

# Shuffle and take 5000 samples
random.shuffle(public_data)
public_data = public_data[:5000]
print(f'Using {len(public_data)} public examples')
```
